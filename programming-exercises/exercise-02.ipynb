```python
# import numpy as np
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/jammext/ds-repository/main/COVID19-Cases-Global.xlsx'
#Question 1: Display the first 10 rows of the data from the given URL.
#Enter your code on this cell and execute it. Please provide a screenshot of your code and result, and submit it to Blackboard.

df = pd.DataFrame(pd.read_excel(url))
display(df.head(10))
```

```python
#Question 2: Display only two countries of your choice. Ensure that it is different from your classmates.
#Please enter your code on this cell and execute it.
#Please provide a screenshot of your code and result, and submit it to Blackboard.

#First letter of Firstname and Surname

#Find a country that starts with the letter 'K' and 'P'     
country_k = df[df['country'].str.startswith('K')]
country_p = df[df['country'].str.startswith('P')]

#Print search results
# print('Countries starting with K and P:', end='')
# display(country_k['country'].unique())
# display(country_p['country'].unique())
# print('—'*50, end='')

# Choose Kosovo and Portugal
dfKosovo = country_k[country_k['country'] == 'Kosovo']
dfPortugal = country_p[country_p['country'] == 'Portugal']

# Print chosen countries
print('\n\nChosen Country with K:', end='')
display(dfKosovo)
print('—'*50, end='')


print('\n\nChosen Country with P:', end='')
display(dfPortugal)
print('—'*50, end='')

```

```python
#Question 3: Display only data from Feb to Nov 2020 of the countries of your choice. Replace any NaNs on the data with zeros.
#Please enter your code on this cell and execute it.
#Please provide a screenshot of your code and result, and submit it to Blackboard.

def dateRange(df):
  return (df['date'] >= '2020-02-01') & (df['date'] <= '2020-11-30')

# Filter data for chosen countries
dfKosovo_filtered = dfKosovo[dateRange(dfKosovo)].fillna(0)
display(dfKosovo_filtered)

dfPortugal_filtered = dfPortugal[dateRange(dfPortugal)].fillna(0)
display(dfPortugal_filtered) 
```

```python
#Question 4: Create another column in the dataset. The new column is the Case Fatality Rate or CFR. 
#It is calculated as the ratio of the total deaths against the total cases.
#Please enter your code on this cell and execute it.
#Please provide a screenshot of your code and result, and submit it to Blackboard.

df = df.fillna(0)
df['CFR'] = (df['total_deaths'] / df['total_cases']).fillna(0)
display(df)
```

```python
#Question 5: Further arrange your dataset to be similar to the output (screenshot) shown on Blackboard.
#Please enter your code on this cell and execute it.
#Please provide a screenshot of your code and result, and submit it to Blackboard.

dfPhilippines = df[df['country'] == 'Philippines'][['country', 'date', 'CFR']]
dfPhilippines = dfPhilippines[dateRange(dfPhilippines)]

dfYemen = df[df['country'] == 'Yemen'][['country', 'date', 'CFR']]
dfYemen = dfYemen[dateRange(dfYemen)]

# display(dfPhilippines.head(10))
# display(dfYemen.head(10))

dfPhilippinesYemen = pd.merge(dfPhilippines, dfYemen, on='date', suffixes=('1', '2'))
dfPhilippinesYemen = dfPhilippinesYemen.iloc[:, [1,0,2,3,4]]
dfPhilippinesYemen['Abs. Diff'] = abs(dfPhilippinesYemen['CFR1'] - dfPhilippinesYemen['CFR2'])
display(dfPhilippinesYemen)
```
